```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import seaborn as sns
```

```python
data = pd.read_csv("News_Category_Dataset_with_splits.csv")
```

```python
#Category Distribution
category_counts = data['category'].value_counts()
plt.figure(figsize=(6, 4))
sns.barplot(x=category_counts.index, y=category_counts.values, palette=[(1,0.99,0.45), (0.98,0.82,0.47), (0.99,0.87,0.89),(0.96,0.49,0.49)])
plt.title("Category Distribution")
plt.xlabel("Category")
plt.ylabel("Count")
plt.xticks(rotation=10)
for i, count in enumerate(category_counts.values):
    plt.text(i, count, str(count), ha='center', va='bottom')
plt.show()
```

```python
# Split Distribution
split_counts = data['split'].value_counts()
sns.barplot(x=split_counts.index, y=split_counts.values)
plt.title("Split Distribution")
plt.xlabel("Split")
plt.ylabel("Count")

for i, count in enumerate(split_counts.values):
    plt.text(i, count, str(count), ha='center', va='bottom')

plt.show()
```

```python
# WordCloud
text = " ".join(data['text'])
wordcloud = WordCloud(width=800, height=400, max_words=200, background_color='white').generate(text)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Text Content Word Cloud")
plt.show()
```

```python
Tech_lengths = data[data['category'] == 'Sci/Tech']['text'].apply(len)
Business_lengths = data[data['category'] == 'Business']['text'].apply(len)
World_lengths = data[data['category'] == 'World']['text'].apply(len)
Sports_lengths = data[data['category'] == 'Sports']['text'].apply(len)

#Text Length Distribution
plt.figure(figsize=(10, 6))
plt.hist(Tech_lengths, bins=50, alpha=0.5, label='Tech News', color='Orange')
plt.hist(Business_lengths, bins=50, alpha=0.5, label='Business News', color='pink')
plt.hist(Sports_lengths, bins=50, alpha=0.5, label='Sports News', color='yellow')
plt.hist(World_lengths, bins=50, alpha=0.5, label='World News', color='red')
plt.title("Text Length Distribution")
plt.xlabel("Text Length")
plt.ylabel("Count")
plt.legend()
plt.show()
```
