```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import seaborn as sns
```

```python
data = pd.read_csv("fake_or_real_news_with_splits.csv")
```

```python
# Category Distribution
category_counts = data['category'].value_counts()
plt.figure(figsize=(3, 4))
sns.barplot(x=category_counts.index, y=category_counts.values)
plt.title("Category Distribution")
plt.xlabel("Category")
plt.ylabel("Count")
plt.xticks(rotation=10)
for i, count in enumerate(category_counts.values):
    plt.text(i, count, str(count), ha='center', va='bottom')
plt.show()
```

```python
# Split Distribution
split_counts = data['split'].value_counts()
sns.barplot(x=split_counts.index, y=split_counts.values)
plt.title("Split Distribution")
plt.xlabel("Split")
plt.ylabel("Count")

for i, count in enumerate(split_counts.values):
    plt.text(i, count, str(count), ha='center', va='bottom')

plt.show()
```

```python
# WordCloud
text = " ".join(data['text'])
wordcloud = WordCloud(width=800, height=400, max_words=200, background_color='white').generate(text)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Text Content Word Cloud")
plt.show()
```

```python
fake_text = data[data['category'] == 'FAKE']['text']

# Word Cloud for 'fake' Category Text
fake_text = " ".join(fake_text)
wordcloud = WordCloud(width=800, height=400, max_words=200, background_color='white').generate(fake_text)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud for 'fake' Category Text")
plt.show()
```

```python
fake_text = data[data['category'] == 'REAL']['text']

# Word Cloud for 'real' Category Text
fake_text = " ".join(fake_text)
wordcloud = WordCloud(width=800, height=400, max_words=200, background_color='white').generate(fake_text)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud for 'real' Category Text")
plt.show()
```

```python
fake_lengths = data[data['category'] == 'FAKE']['text'].apply(len)
real_lengths = data[data['category'] == 'REAL']['text'].apply(len)

#Text Length Distribution
plt.figure(figsize=(10, 6))
plt.hist(fake_lengths, bins=50, alpha=0.5, label='Fake News', color='red')
plt.hist(real_lengths, bins=50, alpha=0.5, label='Real News', color='blue')
plt.title("Text Length Distribution (Fake vs. Real News)")
plt.xlabel("Text Length")
plt.ylabel("Count")
plt.legend()
plt.show()
```
